```python
from datasets import load_dataset
dataset = load_dataset("arxiv_dataset", data_dir="data", ignore_verifications=True)
```

```python
# 데이터셋의 구조 확인
print("Dataset structure:")
print(dataset)

# 첫 몇 개의 샘플 확인
print("\nFirst few samples:")
for i, example in enumerate(dataset['train'].select(range(5))):  # 'train' 대신 실제 스플릿 이름을 사용하세요
    print(f"\nSample {i + 1}:")
    print(f"ID: {example['id']}")
    print(f"Categories: {example['categories']}")
    print(f"Title: {example['title']}")

# categories 필드의 고유한 값들을 확인
print("\nUnique categories:")
unique_categories = set()
for example in dataset['train']:  # 'train' 대신 실제 스플릿 이름을 사용하세요
    if isinstance(example['categories'], str):
        categories = example['categories'].split()  # 공백으로 분리된 경우
    elif isinstance(example['categories'], list):
        categories = example['categories']
    else:
        continue
    unique_categories.update(categories)

for category in sorted(unique_categories):
    print(category)

# 메인 카테고리 분포 확인
print("\nMain category distribution:")
main_category_count = {}
for example in dataset['train']:  # 'train' 대신 실제 스플릿 이름을 사용하세요
    if isinstance(example['categories'], str):
        categories = example['categories'].split()
    elif isinstance(example['categories'], list):
        categories = example['categories']
    else:
        continue
    for category in categories:
        main_cat = category.split('.')[0]
        main_category_count[main_cat] = main_category_count.get(main_cat, 0) + 1

for main_cat, count in sorted(main_category_count.items(), key=lambda x: x[1], reverse=True):
    print(f"{main_cat}: {count}")
```
